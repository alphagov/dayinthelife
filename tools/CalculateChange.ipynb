```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv("with-change.csv", header=None, parse_dates=[0], index_col=[0, 1])
data.columns = ["visits", "proportion", "change"]
```

```python
data["ewma_visits"] = pd.ewma(data, span=14).visits
data["ewma_proportion"] = pd.ewma(data, span=14).proportion
```

```python
shifted = data.unstack().shift(1).stack()
data["prev_ewma_proportion"] = shifted["ewma_proportion"]
data["prev_ewma_visits"] = shifted["ewma_visits"]
```

```python
data
```

```python
step_one = data[data["prev_ewma_visits"] > 100]
step_two = step_one[step_one["visits"] > 100]
```

```python
step_two.change = (step_two.proportion - step_two.prev_ewma_proportion) / step_two.prev_ewma_proportion
```

```python
step_two.to_csv("full.csv")
```

```python
step_two.xs("/when-do-the-clocks-change", level=1).change.plot()
```
